```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report


from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
inputs_n=inputs.drop(['Q1','Q2','Q3','Q4','Q5','Q6','Q7','Q8','Q9','Q10','Q11','Q12','Q13','Q14','Q15','Q16','Q17','Q18','Q19','Q20','Q21','Q22','Q23','Q24','Q25','Q26','Q27','Q28','Q29','Q30','Q31','Q32'],axis='columns')
# inputs_n=inputs.drop(['Q1','Q2','Q3','Q4'],axis='columns')
inputs_n

```

```python

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(inputs_n, target, test_size=0.2, random_state=42)


#Train and evaluate Naive Bayes
nb_classifier = MultinomialNB()
nb_classifier.fit(X_train, y_train)
nb_predictions = nb_classifier.predict(X_test)
nb_accuracy = accuracy_score(y_test, nb_predictions)

print("MNB Naive Bayes Accuracy:", nb_accuracy)
print(classification_report(y_test, nb_predictions))



# Initialize and train the classifier
Classifier = GaussianNB()
Classifier.fit(X_train, y_train)
nb_predictions = Classifier.predict(X_test)
nb_accuracy = accuracy_score(y_test, nb_predictions)

print("GNBNaive Bayes Accuracy:", nb_accuracy)
print(classification_report(y_test, nb_predictions))



#Train and evaluate SVM
svm_classifier= SVC()
svm_classifier.fit(X_train, y_train)
svm_predictions = svm_classifier.predict(X_test)
svm_accuracy = accuracy_score(y_test, svm_predictions)

print("SVM Accuracy:", svm_accuracy)
print(classification_report(y_test, svm_predictions))

# Train and evaluate KNN
knn_classifier = KNeighborsClassifier()
knn_classifier.fit(X_train, y_train)
knn_predictions = knn_classifier.predict(X_test)
knn_accuracy = accuracy_score(y_test, knn_predictions)

print("KNN Accuracy:", knn_accuracy)
print(classification_report(y_test, knn_predictions))
```
