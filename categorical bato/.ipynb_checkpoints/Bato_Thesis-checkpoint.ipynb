```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report


from sklearn.naive_bayes import MultinomialNB
```

```python
#========================== Loading Dataset =======================================>>>
```

```python
data=pd.read_csv('data.csv', encoding='latin-1')
# Drop any rows with missing values

# # Drop irrelevant columns (if any)
# data.drop(columns=['A','B','C','D'], inplace=True)
# data.dropna(inplace=True)


print(data['results'].value_counts())


#Countplot to visualize the distribution of the polarity labels
plt.figure(figsize=(14, 6))
sns.countplot(x='results', data =data, color='orange')
plt.title("Distribution of Results Labels")
plt.show()

# Pie chart to visualize the distribution of polarity labels
plt.figure(figsize=(20, 10))
data['results'].value_counts().plot(kind='pie', autopct = '%1.1f%%')
plt.title("Distribution of results Labels (Pie Chart)")
plt.xlabel("Results")
plt.ylabel("Count")

plt.show()
data
```

```python
#========================== Convert into Numerical Data  =======================================>>>
```

```python
Numerics=LabelEncoder()


inputs= data.drop('prediction',axis='columns')
target= data['prediction']



#creating new dataframe

inputs['E-1']=Numerics.fit_transform(inputs['E'])
inputs['F-1']=Numerics.fit_transform(inputs['F'])
inputs['G-1']=Numerics.fit_transform(inputs['G'])
inputs['H-1']=Numerics.fit_transform(inputs['H'])
inputs['I-1']=Numerics.fit_transform(inputs['I'])
inputs['J-1']=Numerics.fit_transform(inputs['J'])
inputs['K-1']=Numerics.fit_transform(inputs['K'])
inputs['L-1']=Numerics.fit_transform(inputs['L'])
inputs['M-1']=Numerics.fit_transform(inputs['M'])
inputs['N-1']=Numerics.fit_transform(inputs['N'])
inputs['O-1']=Numerics.fit_transform(inputs['O'])
inputs['P-1']=Numerics.fit_transform(inputs['P'])
inputs['Q-1']=Numerics.fit_transform(inputs['Q'])
inputs['R-1']=Numerics.fit_transform(inputs['R'])
inputs['S-1']=Numerics.fit_transform(inputs['S'])
inputs['T-1']=Numerics.fit_transform(inputs['T'])


print(target)
inputs
```

```python
inputs_n=inputs.drop(['E','F','G','H','I','J','K','L','M','N','O','P','Q','R','S','T'],axis='columns')
inputs_n

```

```python

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(inputs_n, target, test_size=0.3, random_state=42)


#Train and evaluate Naive Bayes
nb_classifier = MultinomialNB()
nb_classifier.fit(X_train, y_train)
nb_predictions = nb_classifier.predict(X_test)
nb_accuracy = accuracy_score(y_test, nb_predictions)

print("Naive Bayes Accuracy:", nb_accuracy)
print(classification_report(y_test, nb_predictions))
```

```python
#========================== Training with Balance Data  =======================================>>>
```

```python
from imblearn.over_sampling import SMOTE

#Load the CSV file with polarity labels
input_csv =  "new_data.csv"
data = pd.read_csv(input_csv, encoding='latin-1')

#Calculate the highest number of instanses among polarity
max_instances = data['prediction'].value_counts().max()



# Drop irrelevant columns (if any)
data.drop(columns=['A','B','C','D'], inplace=True)
# data.dropna(inplace=True)
print(data['prediction'].value_counts())

Numerics=LabelEncoder()
inputs= data.drop('prediction',axis='columns')
target= data['prediction']
target
data
```

```python
#creating new dataframe

inputs['E-1']=Numerics.fit_transform(inputs['E'])
inputs['F-1']=Numerics.fit_transform(inputs['F'])
inputs['G-1']=Numerics.fit_transform(inputs['G'])
inputs['H-1']=Numerics.fit_transform(inputs['H'])
inputs['I-1']=Numerics.fit_transform(inputs['I'])
inputs['J-1']=Numerics.fit_transform(inputs['J'])
inputs['K-1']=Numerics.fit_transform(inputs['K'])
inputs['L-1']=Numerics.fit_transform(inputs['L'])
inputs['M-1']=Numerics.fit_transform(inputs['M'])
inputs['N-1']=Numerics.fit_transform(inputs['N'])
inputs['O-1']=Numerics.fit_transform(inputs['O'])
inputs['P-1']=Numerics.fit_transform(inputs['P'])
inputs['Q-1']=Numerics.fit_transform(inputs['Q'])
inputs['R-1']=Numerics.fit_transform(inputs['R'])
inputs['S-1']=Numerics.fit_transform(inputs['S'])
inputs['T-1']=Numerics.fit_transform(inputs['T'])


#inputs
inputs_n=inputs.drop(['E','F','G','H','I','J','K','L','M','N','O','P','Q','R','S','T'],axis='columns')
inputs_n




#Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(inputs_n, target, test_size=0.2, random_state= 42)



# Oversample the minority classes using SMOTE with reduced number of neighbors
smote = SMOTE(sampling_strategy={
    'middle income mid': max_instances,
    'middle income low': max_instances,
    'low income low': max_instances,
    'low income poor': max_instances,
    'middle income high': max_instances
    
}, random_state=42, k_neighbors=2)  # Adjust the value of k_neighbors

# Resample the training data
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)





#Create a data frame from the resampled data
resampled_data = pd.DataFrame({'prediction': y_resampled})

# #Countplot to visualize the distribution of polarity labels after oversampling





plt.figure(figsize=(12, 15))

# Countplot to visualize the distribution of polarity labels before oversampling
plt.subplot(2, 1, 1)
sns.countplot(x='prediction', data=data, color='orange')
plt.title("Distribution of Prediction Labels (Before Oversampling)")

# Countplot to visualize the distribution of polarity labels after oversampling
plt.subplot(2, 1, 2)
sns.countplot(x='prediction', data=resampled_data, color='orange')
plt.title("Distribution of Prediction Labels (After Oversampling)")

# Display counts of each class in the resampled data
print("Counts of each class in resampled data:")
print(resampled_data['prediction'].value_counts())



plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

```python
#Naive bayes in Balance Data Training

#Train and evaluate Naive Bayes
nb_classifier = MultinomialNB()
nb_classifier.fit(X_resampled, y_resampled)
nb_predictions = nb_classifier.predict(X_test)
nb_accuracy = accuracy_score(y_test, nb_predictions)
print("Naive Bayes Accuracy:", nb_accuracy)
print(classification_report(y_test, nb_predictions))

```
